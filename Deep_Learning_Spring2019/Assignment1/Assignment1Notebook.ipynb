```python
import numpy as np
import copy
import matplotlib.pyplot as plt
mean = [30, 30]
cov = [[5,0],[0,5]]
x, y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x, y, 'x')
plt.axis('equal')
data1 = []
for i in range(100):
    data1.append([[x[i],y[i]],1])
mean = [15, 15]
cov = [[5,0],[0,5]]
x, y = np.random.multivariate_normal(mean, cov, 100).T
data2 = []
for i in range(100):
    data2.append([[x[i],y[i]],-1])
dataTr = copy.copy(data1)
dataTr += data2
np.random.shuffle(dataTr)
dataTest = dataTr[:40]
dataTr = dataTr[40:]
x1=[]
y1=[]

for i in range(len(dataTr)):
    x1.append(dataTr[i][0][0])
    y1.append(dataTr[i][0][1])
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
plt.plot(x1, y1, 'x')
plt.axis('equal')
plt.show()
```

```python
import perceptron as pt
_perceptron = pt.Perceptron(2)
vectors = []
labels = []
for i in range(len(dataTr)):
    vectors.append(dataTr[0])
    labels.append(dataTr[1])
for i in range(500):
    _perceptron.train(vectors, labels)
    
print(_perceptron.w)
accuracy = 0
for data in dataTr:
    result = _perceptron.forward(np.array(data[0]))
    if(result == data[1]):
        accuracy+=1
print(accuracy/len(dataTr))
accuracy=0
for data in dataTest:
    result = _perceptron.forward(np.array(data[0]))
    if(result == data[1]):
        accuracy+=1
print(accuracy/len(dataTest))
    #print(data)
    #print(data, _perceptron.forward(np.array(data)))
```
